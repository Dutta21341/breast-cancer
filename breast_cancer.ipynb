IMPORTING NECESSARY LIBRARIES

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import sklearn.datasets
```

LOADING THE IN-BUILT DATASET IN SCIKIT-LEARN LIBRARY

```python
breast_cancer_ds = sklearn.datasets.load_breast_cancer()
print(breast_cancer_ds)
```

CONVERTED TO DATAFRAME USING PANDAS LIBRARY

```python
df = pd.DataFrame(breast_cancer_ds.data, columns = breast_cancer_ds.feature_names)
```

```python
df.head()
```

```python
df['label']  =breast_cancer_ds.target
```

```python
X = df.drop(columns = 'label',axis=1)
Y = df['label']
```

```python
print(X,Y)
```

SPLITTING THE DATASET INTO TRAIN AND TEST

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, stratify = Y, random_state = 0)
```

Neural Network

```python
#importing tensorflow and keras
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
model = keras.Sequential([
    tf.keras.layers.Flatten(input_shape = (30,)),
    tf.keras.layers.Dense(20, activation = 'relu'),
    tf.keras.layers.Dense(2, activation = 'sigmoid')])
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics= ['accuracy'])
```

```python
history = model.fit(X_train, Y_train, validation_split=0.1, epochs=10)
```

```python
#standardize the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.fit_transform(X_test)
```

```python
print(X_train_std)
print(X_test_std)
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')

plt.xlabel('epoch')
plt.legend(['training_data', 'validation_data'], loc = 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')

plt.xlabel('epoch')
plt.legend(['training_data', 'validation_data'], loc = 'upper right')
```

```python
loss, accuracy = model.evaluate(X_test_std, Y_test)
```
